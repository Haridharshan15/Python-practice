```python
#palindrome number

def pal(n):
  temp=n
  c=0
  b=0
  while temp!=0:
    b=int(temp%10)
    temp=int(temp/10)
    #print(b," ",temp)
    #c=b+c
    c=(c*10)+b
    # print("in loop",c)
  #c=int(c/10)
  return "Pallindrome" if c==n else "Not Pallindrome"

n=1223221
#n=int(input("Enter a number:"))
print(pal(n))
```

```python
#amstrong number
def fib(n):
  temp=n
  c=0
  while temp!=0:
    b=int(temp%10)
    temp=int(temp/10)
    c=c+(b*b*b)
    b=0
  return "Amstrong" if c==n else "Not Amstrong"

n=153
print(fib(n))
```
