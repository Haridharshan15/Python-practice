```python
#Match-case statement in python(like switch case in c,c++)
b,n,c=input("Enter the number : ").split()
match n:
  case '+':
    print("Addition")
    print(f"{b}-{c} = ",int(b)+int(c))
  case '-':
    print("Subtraction")
    print(f"{b}-{c} = ",int(b)-int(c))
  case '*':
    print("Multiplication")
    print(f"{b}*{c} = ",int(b)*int(c))

  case '/':
    print("Division")
    print(f"{b}/{c} = ",int(b)/int(c))

  case '%':
    print("Modulus")

    print(f"{b}%{c} = ",int(b)%int(c))

  case _:
    print("Invalid input")


```

```python
#property decorator
class hari:
  def __init__(self,name,age):
    self.n=name
    self.k=age
  @property        #property decorator [fuction will be change into property]
  def msg(self):
    return self.n + " is "+str(self.k)+" years old"
x=hari("Dharshan",22)
print(x.n)
print(x.k)
print(x.msg)      # msg is a property so thats why we use here "x.msg" instead of "x.msg()"==> this is for call a function
x.k=30
print(x.msg)
```

```python
#property decorator gettarr setaar
class hari:
  def __init__(self,total): #in python dont have private and puplic variable we assume that
    self._tot=total        #_tot "_" means private dont access directly  like encaptulation
  def avg(self):
    return self._tot/5.0
  @property          #property decorator [fuction will be change into property]
  def tot(self):
    return self._tot
  @tot.setter          # to set the vale for _tot indirectly
  def tot(self,t):
    if t<0 or t>100:      #validation process also we can uese in setter
      print("Invalid Value")
    else:
      self._tot=t
o=hari(100)
print(o.tot)
print(o.avg())

o.tot=90       #to send the another value to _tot through tot
print(o.tot)
print(o.avg())

o.tot=110       #to send the another value to _tot through tot
print(o.tot)
print(o.avg())
```

```python
#property method
class hari:
  def __init__(self,total): #in python dont have private and puplic variable we assume that
    self._tot=total        #_tot "_" means private dont access directly  like encaptulation
  def avg(self):
    return self._tot/5.0
                        # instead of property decorator,we use down tot=property(tot,tot)
  def tot(self):
    return self._tot

  def setter(self,t):     # to set the vale for _tot indirectly
    if t<0 or t>100:      #validation process also we can uese in setter
      print("Invalid Value")
    else:
      self._tot=t
  tot=property(tot,setter) #instead of "@decorator" we use this
o=hari(100)
print(o.tot)
print(o.avg())

o.tot=200
print(o.tot)
print(o.avg())
```

```python
#classmethod
class Muthu:
  count=0
  def __init__(self,n,a):
    self.name=n
    self.age=a
    Muthu.count+=1

  def printDetail(self):
    print(f"My name {self.name} My age is {self.age}")
  @classmethod         #classmethod
  def total(cls):
    return cls.count

o=Muthu("Hari",23)
o.printDetail()

print("Total count : ",Muthu.total()) #same as "o.total()"

ab=Muthu("Hari",23)
ab.printDetail()

print("Total count : ",o.total())
```

```python
#inheritance
#==>single inheritance for example with passing arguments
class Father:
  def __init__(self,name):
    self.name=name

  def father(self):
    print(f"My name is {self.name}")
    print("Parent class executed")

class Son(Father):
  def __init__(self,name,age):
    self.age=age
    Father.__init__(self,name)
  def son(self):
    print(f"My name is {self.name}")
    print(f"My age is {self.age}")

x=son("Hari",23)
x.father()
x.son()

```

```python
#Multileval inheritance
#A class is derived from another derived class (like a chain).
class GrandParent:
    def feature1(self):
        print("Feature 1")

class Parent(GrandParent):
    def feature2(self):
        print("Feature 2")

class Child(Parent):
    def feature3(self):
        print("Feature 3")

obj = Child()
obj.feature1()
obj.feature2()
obj.feature3()
```

```python
#inheritance
#==>hierarical inheritance ==many child to one parent
# Hybrid (multiple + multilevel),Combination of two or more types (Single + Multiple + Hierarchical).
#in this we have to create a seperate object
class father:
  def house(self):
    print("House")

class mother:
  def Gold(self):
    print("Gold")

class son(father): #hierarical= "daughter" also inherit from father
  def car(self):
    print("Car")

class daughter(father):   #hierarical="son" also inherit from father
  def scooty(self):
    print("Scooty")

class son1(father,mother):   #here both # Hybrid (multiple + multilevel)
  def bike(self):
    print("bike")

x=son() # Create an instance of son
x.house()
x.car()
print("---------------")
y=daughter() # Create an instance of daughter
y.scooty()
y.house()
# daughter does not inherit from mother, so y.Gold() will cause an error
# y.Gold()
print("---------------")
z=son1() # Create an instance of son1
z.bike()
z.house()
z.Gold()
```
